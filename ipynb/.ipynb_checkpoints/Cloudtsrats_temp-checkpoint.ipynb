# Calculating beta

```python
df_complete = pd.read_csv(r"D:\Cloudstrats\Tasks\Task #4 - Power BI\NSE-50 Project\NIFTY50_index_merged.csv")
df_complete.head()
```

```python
df_complete.info()
```

```python
df_complete['Symbol'].value_counts()
```

```python
grouped_data = df_complete.groupby('Symbol')
grouped_df = [grouped_data.get_group(x) for x in grouped_data.groups]
len(grouped_df)
grouped_df_list = []

for group_df in grouped_df:
    
    # print("1 = ", group_df.rolling(1).var(group_df['NSE_Index_pct_change']))
    # print(group_df['NSE_Index_pct_change'].rolling(2).var())
    # print(group_df['Symbol_pct_change'].rolling(2).cov(group_df['NSE_Index_pct_change']))
    group_df['beta'] = group_df['Symbol_pct_change'].rolling(2).cov(group_df['NSE_Index_pct_change']) / group_df['NSE_Index_pct_change'].rolling(2).var()
    grouped_df_list.append(group_df)
```

```python
count = 0
first_one = None
for one_var in grouped_df_list:
    if count==0:
        first_one = one_var
        count += 1
    else:
        first_one = first_one.append(one_var, ignore_index=True)
        # first_one = pd.merge(first_one, one_var, how='left', on='Date')
```
